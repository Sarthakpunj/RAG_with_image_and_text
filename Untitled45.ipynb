```python
def get_image(url, filename, extension):
    # Download image content from the URL
    content = requests.get(url).content

    # Define the full path
    file_path = f'/content/{filename}.{extension}'

    # Save the image to disk
    with open(file_path, 'wb') as f:
        f.write(content)

    # Open and show the image
    image = Image.open(file_path)
    image.show()

    # Return the image object
    return image
```

```python
loader = TextLoader("/content/nike.txt")
print(loader.load()[0].page_content)
```

```python
data=loader.load()[0].page_content
```

```python
def get_text_chunks(text):
  text_spliter = CharacterTextSplitter(chunk_size = 20,chunk_overlap=10)
  docs = [Document(page_content=x) for x in text_spliter.split_text(text)]
  return docs
```

```python
docs = get_text_chunks(data)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model = "models/embedding-001")
```

```python
vectorstore = FAISS.from_documents(docs,embedding=embeddings)
```

```python
retriever=vectorstore.as_retriever()
```

```python
retriever.invoke("Nike")
```

```python
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
```

```python
llm_vision = laod_model("gemini-2.0-flash")
```

```python
llm_text = laod_model("gemini-2.0-flash")
```

```python
template = """
'''
{context}
'''

{information}

provide brief information.

"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
rag_chain = (
    {"context":retriever,"information":RunnablePassthrough()}
    | prompt
    | llm_text
    | StrOutputParser()

)
```

```python
result = rag_chain.invoke("give me details on Nike")
```

```python
result
```

```python
full_chain = (
    RunnablePassthrough() | llm_vision | StrOutputParser() |rag_chain
)
```
